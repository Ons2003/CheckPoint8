```python
#4
f = open(path)
def get_number_of_words(your_file) :
  words = list(your_file)
  number_of_words = 0

  for x in words : 
    i = x.split()
    number_of_words = number_of_words + len(i)
  
  return number_of_words 

get_number_of_words(f)

```
